```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
g = sns.pairplot(iris,hue='species', palette='husl', markers='d', size=2.5, plot_kws=
    {
    "s":40,
    "alpha":1.0,
    'lw':0.5,
    'edgecolor':'k'
    })
#handles = g._legend_data.values()
#labels = g._legend_data.keys()
#g.fig.legend(handles=handles, labels=labels, loc='lower center', ncol=3, title='abc')
#g._legend.set_bbox_to_anchor((0.5, 0.5))
#g.fig.subplots_adjust(bottom=0.13)
sns.move_legend(g, bbox_to_anchor=(0.5, -0.09), loc="lower center", ncol=3, labelspacing=0.8, fontsize=18, title='Samples', title_fontsize='18')
plt.tight_layout()
```
